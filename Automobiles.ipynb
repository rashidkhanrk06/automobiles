#  **Exploratory Data Analysis (EDA) of Car Data** 




### Introduction 

The Automobiles Dataset is a comprehensive collection of data pertaining to various attributes of automobiles. It provides valuable insights into the features and specifications of different cars, facilitating analysis and understanding of the automotive industry. The dataset is compiled to assist researchers, analysts, and enthusiasts in exploring patterns, correlations, and trends within the automotive domain. The data encompasses a wide range of characteristics, allowing for a detailed examination of factors influencing automobile performance, cost, and market dynamics.

### Features of the Dataset

1. **Symboling**: Represents the insurance risk rating of the vehicle, which provides an assessment of its safety and risk.

2. **Normalized-losses**: Denotes the normalized losses in terms of insurance claims for each vehicle, providing insights into their risk factor.

3. **Make**: Specifies the manufacturer or brand of the automobile, indicating the company responsible for its production.

4. **Fuel-type**: Indicates the type of fuel used by the vehicle, such as gasoline or diesel.

5. **Aspiration**: Represents whether the automobile is naturally aspirated or equipped with forced induction.

6. **Num-of-doors**: Specifies the number of doors on the vehicle, a feature significant for both functionality and aesthetics.

7. **Body-style**: Describes the general shape and style of the vehicle, such as sedan, hatchback, coupe, etc.

8. **Drive-wheels**: Indicates the type of wheels that provide power to the vehicle, influencing its handling and performance.

9. **Engine-location**: Identifies the placement of the engine, which can be either front or rear, impacting the vehicle's weight distribution and handling.

10. **Wheel-base, Length, Width, Height**: These dimensions provide physical measurements of the vehicle, aiding in assessing its size and proportions.

11. **Curb-weight**: Represents the weight of the vehicle without occupants or cargo, a crucial factor in determining its performance and fuel efficiency.

12. **Engine-type**: Denotes the type of engine used in the vehicle, such as 'dohc', 'ohcv', etc.

13. **Num-of-cylinders**: Specifies the number of cylinders in the vehicle's engine, influencing its power and efficiency.

14. **Engine-size**: Represents the volume of the engine's combustion chambers, affecting the power and performance of the vehicle.

15. **Fuel-system**: Describes the type of fuel injection system in the vehicle, which impacts its fuel delivery and efficiency.

16. **Bore, Stroke, Compression-ratio**: These parameters pertain to the engine's internal combustion characteristics, affecting its efficiency and power output.

17. **Horsepower, Peak-rpm**: Indicate the engine's horsepower and peak revolutions per minute, providing information about its power and performance.

18. **City-mpg, Highway-mpg**: Represent the miles per gallon (mpg) of fuel efficiency in city and highway driving, respectively.

19. **Price**: Specifies the cost of the automobile, a crucial factor for market analysis and consumer decision-making.

Understanding and analyzing these features can yield valuable insights into the automotive industry, helping in predicting vehicle performance, assessing market trends, and making informed business and consumer decisions.

### Target Variable Analysis

```python
cars.describe()['price'].plot(kind='bar', figsize=(5,4) )
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(11,4))

sns.boxplot(data=cars, x='price', ax=axs[0])
sns.histplot(data=cars, x='price', ax=axs[1], bins=25 )
plt.show()
```

### Inference Statement:
The visualizations generated from the code provide valuable insights into the distribution of 'price' in the automobiles dataset.

- **Boxplot (Left Plot)**:
  - The boxplot indicates that the 'price' distribution has a median around a certain value, which is represented by the line within the box. The interquartile range (IQR) gives an idea of the spread of the prices. The presence of any outliers beyond the whiskers can suggest potential extreme values or anomalies in the 'price' data.

- **Histogram (Right Plot)**:
  - The histogram illustrates the frequency distribution of 'price'. It shows that a significant number of automobiles fall within a certain price range, with a peak in the histogram indicating a common price range. The shape of the histogram can provide insights into the overall distribution, whether it is skewed, bimodal, or normally distributed.

In summary, the visualizations demonstrate that the majority of automobiles in the dataset are concentrated within a specific price range, as indicated by the peak in the histogram. Additionally, the boxplot highlights the spread of prices and identifies any potential outliers that may require further investigation. These visualizations aid in understanding the distribution and variability of prices in the dataset, which is crucial for pricing analysis and decision-making in the automotive industry.

##### Correlation with Target variable

```python
cor_matrix=cars.corr()['price']
cor_matrix=cor_matrix[abs(cor_matrix)>.5]
cor_matrix.sort_values(ascending=False)
```

```python
matrix=cars[cor_matrix.index].corr()
sns.heatmap(matrix, cmap='Paired',
           annot = True, annot_kws={'size': 6},
            square=True, linecolor='w',
            linewidths=0.1) 
plt.show()
```

### Inference Statement:
The visual representation of the correlation matrix through the heatmap provides valuable insights into the relationships between different features and the 'price' of automobiles in the dataset.

- **Strong Positive Correlations**:
  - The heatmap shows that certain features have a strong positive correlation with the 'price' of the automobiles. Features like 'engine-size' and 'horsepower' exhibit strong positive correlations, suggesting that as these features increase, the 'price' of the automobiles tends to increase as well.

- **Moderate Positive Correlations**:
  - Additionally, features like 'curb-weight', 'width', and 'length' also demonstrate moderate positive correlations with the 'price', indicating that as these features increase, the 'price' tends to increase as well, though not as strongly as the highly correlated features.

- **No Negative Correlations**:
  - Notably, there are no strong negative correlations present in the heatmap, suggesting that none of the features have a significant negative linear relationship with the 'price'. This indicates that as the features do not decrease, the 'price' does not significantly decrease either.


```python
columns=[['curb-weight','width','engine-size'],
         ['length','horsepower','wheel-base']]
fig,axes=plt.subplots(2,3,figsize=(11,10))
for col_name, axs in zip(columns,axes):
    for x,ax in zip(col_name,axs):
        g=sns.scatterplot(data=cars, x=x, y='price', ax=ax)
        g.set_title(f'+ve Corr b/w {x} and Price')
plt.show()
```

### Inference Statement:
The scatter plots produced by the code offer valuable insights into the relationships between specific features and the 'price' of automobiles in the dataset.

- **Set 1: 'curb-weight', 'width', 'engine-size'**:
  - Scatter plots for 'curb-weight', 'width', and 'engine-size' against 'price' display positive correlations. As 'curb-weight', 'width', and 'engine-size' increase, there is a tendency for the 'price' of the automobiles to also increase. This suggests that heavier cars ('curb-weight'), wider cars ('width'), and cars with larger engines ('engine-size') tend to be associated with higher prices.

- **Set 2: 'length', 'horsepower', 'wheel-base'**:
  - Scatter plots for 'length', 'horsepower', and 'wheel-base' against 'price' similarly demonstrate positive correlations. An increase in 'length', 'horsepower', and 'wheel-base' is associated with a tendency for the 'price' of the automobiles to increase. This implies that longer cars ('length'), cars with higher horsepower ('horsepower'), and cars with longer wheelbases ('wheel-base') tend to have higher prices.

```python
columns=['city-mpg','highway-mpg']
fig,axs=plt.subplots(1,2,figsize=(8,4))
for col, ax in zip(columns,axs):
    g=sns.lineplot(data=cars, x=col, y='price', ax=ax)
    g.set_title(f'-ve Corr b/w {x} and Price')
plt.show()   
```

### Inference Statement:
The line plots produced by the code provide valuable insights into the relationships between fuel efficiency (measured in miles per gallon, mpg) and the 'price' of automobiles in the dataset.

- **'city-mpg' and 'highway-mpg'**:
  - The line plots demonstrate a negative correlation between both 'city-mpg' and 'highway-mpg' with the 'price' of the automobiles. As the fuel efficiency in terms of miles per gallon (mpg) increases (indicating more miles covered per gallon of fuel), the 'price' of the automobiles tends to decrease.
  
  - This suggests that cars with higher fuel efficiency, whether in city or highway conditions, are associated with lower prices. Consumers looking for more fuel-efficient options may find cars with higher 'city-mpg' and 'highway-mpg' to be more affordable.

**Univariate Analysis Categorical Variables**

```python
fig,axs=plt.subplots(1,2,figsize=(14,4))
g=sns.countplot(data=cars,x='make', ax=axs[0])
g.set_xticklabels(g.get_xticklabels(),rotation=90)
g.set_title('**Univariate Analysis**')

g=sns.barplot(data=cars,x='make', y='price', ax=axs[1],ci=None)
g.set_xticklabels(g.get_xticklabels(),rotation=90)
g.set_title('**Bivariate Analysis**')

plt.show()
```

Some car manufacturers may have higher average prices, while others may have lower average prices. This indicates that the 'make' of the car can be a significant factor influencing the pricing.
