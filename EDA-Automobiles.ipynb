### Introduction 

The Automobiles Dataset is a comprehensive collection of data pertaining to various attributes of automobiles. It provides valuable insights into the features and specifications of different cars, facilitating analysis and understanding of the automotive industry. The dataset is compiled to assist researchers, analysts, and enthusiasts in exploring patterns, correlations, and trends within the automotive domain. The data encompasses a wide range of characteristics, allowing for a detailed examination of factors influencing automobile performance, cost, and market dynamics.

### Features of the Dataset

1. **Symboling**: Represents the insurance risk rating of the vehicle, which provides an assessment of its safety and risk.

2. **Normalized-losses**: Denotes the normalized losses in terms of insurance claims for each vehicle, providing insights into their risk factor.

3. **Make**: Specifies the manufacturer or brand of the automobile, indicating the company responsible for its production.

4. **Fuel-type**: Indicates the type of fuel used by the vehicle, such as gasoline or diesel.

5. **Aspiration**: Represents whether the automobile is naturally aspirated or equipped with forced induction.

6. **Num-of-doors**: Specifies the number of doors on the vehicle, a feature significant for both functionality and aesthetics.

7. **Body-style**: Describes the general shape and style of the vehicle, such as sedan, hatchback, coupe, etc.

8. **Drive-wheels**: Indicates the type of wheels that provide power to the vehicle, influencing its handling and performance.

9. **Engine-location**: Identifies the placement of the engine, which can be either front or rear, impacting the vehicle's weight distribution and handling.

10. **Wheel-base, Length, Width, Height**: These dimensions provide physical measurements of the vehicle, aiding in assessing its size and proportions.

11. **Curb-weight**: Represents the weight of the vehicle without occupants or cargo, a crucial factor in determining its performance and fuel efficiency.

12. **Engine-type**: Denotes the type of engine used in the vehicle, such as 'dohc', 'ohcv', etc.

13. **Num-of-cylinders**: Specifies the number of cylinders in the vehicle's engine, influencing its power and efficiency.

14. **Engine-size**: Represents the volume of the engine's combustion chambers, affecting the power and performance of the vehicle.

15. **Fuel-system**: Describes the type of fuel injection system in the vehicle, which impacts its fuel delivery and efficiency.

16. **Bore, Stroke, Compression-ratio**: These parameters pertain to the engine's internal combustion characteristics, affecting its efficiency and power output.

17. **Horsepower, Peak-rpm**: Indicate the engine's horsepower and peak revolutions per minute, providing information about its power and performance.

18. **City-mpg, Highway-mpg**: Represent the miles per gallon (mpg) of fuel efficiency in city and highway driving, respectively.

19. **Price**: Specifies the cost of the automobile, a crucial factor for market analysis and consumer decision-making.

Understanding and analyzing these features can yield valuable insights into the automotive industry, helping in predicting vehicle performance, assessing market trends, and making informed business and consumer decisions.

## Relationship between Numeric Features with `Price`

```python
cars.corr()
```

#### Relationship with Target variable

```python
cars.corr()['price']
```

```python
corr_mat = cars.corr()['price'][abs(cars.corr()['price'])>.7]
corr_df =pd.DataFrame(data=corr_mat
                       .values,
                      index=corr_mat
                      .index,
                      columns=['corr_values'])
corr_df                     
```

```python
g=corr_df.plot(kind='barh')
g.set_title('Price corelation')
```

```python
sns.heatmap(cars[corr_df.index].corr(), 
            annot=True,
            cmap='RdBu', 
            center =0,
            vmin=-1,
            vmax=1,
            linewidths=1
                   )
plt.show()
```

### Conclusion
From the above heatmap, `wheel-base`, `length`,`width`,`curb-weight`,`engine-size`and `horsepower` are highly **+ve** correlated with `price`.<br>
and `highway-mpg` is highly **-ve** correlated with `price`.

```python
g=sns.jointplot(x=cars['wheel-base'],
              y=cars['price'],
              ylim=(0,40000))
```

### Conclusion
The above plot clearly shows a linear relationship between â€˜priceâ€™ and â€˜wheel-baseâ€™. The â€˜priceâ€™ increases with an increase in â€˜wheel-baseâ€™.

## Relationship Pattern between Categorical Features and `Price`

```python
sns.countplot(y='make', 
              data=cars,
              order= cars.make
              .value_counts()
              .index )
plt.show()
```

### Conclusion
Referencing the plot above, it is evident that Toyota, Nissan, Honda, Subaru, and Mazda are the top five car manufacturers producing the highest number of cars.

```python
g=sns.catplot(kind='bar', 
              y='make',
              x='price', 
              data=cars,
              order = cars.groupby('make')['price']
              .mean().sort_values(ascending=False).index,
              ci=None)
g.set_xticklabels(rotation=90)
plt.show()
```

```python
g=sns.catplot(data=cars,
           kind='strip', 
           x='make',
           y='price',
           jitter=False,
           palette='Paired',
           order=cars.groupby('make')['price']
           .mean().sort_values(ascending=False).index,
           alpha=1)
g.set_xticklabels(rotation=90) 
```

### Conclusion
Upon observing the above plots, it is evident that, although Jaguar, Mercedes, Porsche, BMW, and Audi are the top five car manufacturers that produce fewer cars but with the higher prices.

```python
g=sns.JointGrid()
sns.barplot(data=cars, x='fuel-type',
            y='price',
            hue='make',
            ax=g.ax_joint,
            ci=None)
sns.histplot(data=cars, x='fuel-type',
            ax=g.ax_marg_x)
sns.histplot(data=cars, y='price',
            ax=g.ax_marg_y)

g.ax_joint.legend(bbox_to_anchor=(1.25, .99), loc='upper left')
```

We can observe a trend: most manufacturers produce cars with gasoline fuel type, while fewer produce diesel cars. The latter tend to have slightly higher prices, although it's not very significant.
